```python

def f_limpieza(mensaje):
    ListaMensaje = mensaje.splitlines()
    Resultado = []
    for linea in ListaMensaje: 
      if linea[0:1].isalpha():
        elemento = linea.replace("-",",")
        elemento = elemento.split(",")  
        nombre = elemento[0]
        id_cliente = elemento[1].replace(" ","")
        id_cliente = id_cliente[0:5]
        id_canal = elemento[1][6:9]
        id_evento = elemento[1][9:14]
        precio = elemento[2]
        Resultado.append([nombre,id_cliente,id_canal,id_evento,precio])
    df = pd.DataFrame(Resultado,columns=['nombre','id','canal','evento','precio'])
    df =  df.sort_values('id')
    return df

def f_selecciona_nombre_mas_comun(df: pd.DataFrame):
  df['moda'] = df.groupby('id').nombre.transform(
    lambda x: pd.Series.mode(x)[0])
  return df  # Df ya procesado
```
